Importing the libraires

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#importing the data
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

```python
credit_card_data.head()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
#checking the class distribution
credit_card_data["Class"].value_counts()
```

From the above we can see this is a highly unbalanced dataset therefore we will perform undersampling teqnuie

note -->

0 - Normal transaction

1 - Fraud transaction

```python
# sperating the data

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print (legit.shape)
print(fraud.shape)
```

```python
#data visualizing

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby("Class").mean()
```

###UnderSampling

number of fraud transactions 492



```python
legit_sample = legit.sample(n=492)   #we are ramdomly sampling 492 samples fromthe legit data
```

```python
#concadinating 2 dataframes

new_dataframe = pd.concat([legit_sample, fraud],axis = 0)
```

```python
new_dataframe.info()
```

```python
new_dataframe["Class"].value_counts()
```

```python
new_dataframe.groupby("Class").mean()
```

Spliting the data into features and target

```python
X = new_dataframe.drop(columns = "Class", axis = 1)
Y = new_dataframe["Class"]
```

```python
X
```

```python
Y
```

Split data into train & test

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training


```python
#logistic Reg

model = LogisticRegression()
```

```python
#traing the model
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#Accuracy score

#training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)


print("Traing data Accuracy: ",training_data_accuracy)
```

```python
#Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)


print("Test data Accuracy: ",test_data_accuracy)
```
